Case Study – Text Analyzer System
Instructions:

Take a string input from the user using input().

Convert the text into lowercase using lower().

Use a for loop to iterate through each character in the string.

Count: • Total vowels
• Total consonants
• Total digits
• Total special characters

Use: • isalpha() for alphabets
• isdigit() for numbers

Display all counts using print()

```python
text = input("Enter your String: ").lower()

v = c = d = sc = 0

for letter in text:

  if letter in "aeiou":
    v += 1
  elif letter.isalpha():
    c += 1
  elif letter.isdigit():
    d +=1
  else:
    sc +=1

print(f"total no of vowels: {v}")
print(f"total no of consonants: {c}")
print(f"total no of digit: {d}")
print(f"total no of sc: {sc}")


```


Case Study – Student Performance Management System
Instructions

Create two lists: students marks

Use range() to: Access and print student names and marks using index positions

Use enumerate() to: Display student index number and student name

Use zip() to: Combine student names and marks together

Display all outputs using for loops

Use proper headings with print() statements

Print the final result in a clean format

```python
#case study

students = ["Minu","Harini","Nandhini","Nivetha"]
marks = [85,98,87,80]

for i in range(len(students)):
  print(f"Student: {students[i]} | Marks: {marks[i]}")

for index, student in enumerate(students):
    print(f"Index: {index} | Student Name: {student}")

for student, mark in zip(students, marks):
    print(f"Student: {student} | Marks: {mark}")

for student, mark in zip(students, marks):
    print(f"{student:<10} : {mark} Marks")

print("\nTotal Students:", len(students))
```
