Case Study: Bank Loan Eligibility System
Instructions:

Create variables: customer_name, age, income, credit_score, existing_loan

Use conditional statements (if, elif, else) to: • Approve loan if age ≥ 21, income ≥ 30000, credit_score ≥ 700, and no existing loan
• Conditionally approve loan if credit_score ≥ 600 and income ≥ 40000
• Otherwise reject the loan application

Display customer details and loan status using print()

```python
#variables

Customer_name =input("Enter Your Name:")
Age = int(input("Enter Your Age:"))
Income = float(input("Enter Your Income:"))
Credit_score = int(input("Enter your Credit_score:"))

Existing_loan = bool(input("Exisitng_loan status:"))

if Age >= 21 and Income >= 30000:
  if Credit_score >= 700 and not Existing_loan:
    status = "Approved"

  elif Credit_score >= 600 and Income >= 40000:
    status = "Conditionally Approved"

  else:
    status = "Not Approved"

else:
  status = "Not Eligible"


print("Loan Status:", status)
```

## **Case Study Scholarship Eligibility Checker**

**Instructions:**

1. Create variables: `student_name`, `marks`, `attendance`, `family_income`.
2. Use conditional statements (`if`, `elif`, `else`) to:

   * Award **Full Scholarship** if marks ≥ 90 and attendance ≥ 90.
   * Award **Half Scholarship** if marks ≥ 80 or attendance ≥ 85.
   * Award **Need-Based Scholarship** if marks ≥ 70 and `family_income` < 30000.
   * Otherwise, display **Not Eligible**.
3. Display all data (no loops).


```python
#creating variables

student_name = input("Enter your name: ")
marks = int(input("Enter your marks: "))
attendance = int(input("Enter your attendance percent: "))
family_income = float(input("Enter your family_income: "))

if marks >= 90 and attendance >= 90:
  scholarship_status = "100%-Scholarship"

elif marks >= 80 and attendance >= 85:
    scholarship_status = "50%-scholarship"

elif marks >= 70 or family_income < 30000:
    scholarship_status = "Need-based Scholarship"

else:
  scholarship_status = "Not-Eligible"




print("Scholarship_status:", scholarship_status)
```
