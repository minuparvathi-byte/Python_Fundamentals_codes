## **Case Study – Student Data Management System**

**Instructions:**

1. Create a dictionary named `student` with the following keys:
   name, age, course

2. Store multiple students' data using lists as values.

3. Add one new student details to the dictionary:

   • Append new name to name list  

   • Append new age to age list  

   • Append new course to course list  

4. Display all student details using dictionary keys.

5. Access and display:

   • First student's details using indexing

   • Newly added student's details using indexing

```python
#Case Study – Student Data Management System


student = {
    "name":["Pavithra","Vaishu","Nivetha","Minu"],
    "age": [33,32,32,33],
    "course":["Full stack","Machine learning","AI","Data analytics"]
    }


#add new student
student["name"].append("Ram")
student["age"].append("34")
student["course"].append("Data science")

#display keys
print("Names:", student["name"])
print("Ages:", student["age"])
print("Courses:", student["course"])


#first student
print("\n---First Student---")
print("Names:", student["name"][0])
print("Ages:", student["age"][0])
print("Courses:", student["course"][0])

#newly added  student
print("\n---New Student---")
print("Names:", student["name"][4])
print("Ages:", student["age"][4])
print("Courses:", student["course"][4])
```
