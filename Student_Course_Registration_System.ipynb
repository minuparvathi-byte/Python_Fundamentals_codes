# Case Study – Student Course Registration System using Set
Instructions:

Create a set named courses with course names: Python, SQL, Power BI, Python, Excel

Perform set operations: • Add course using add()
• Add multiple courses using update()
• Remove course using remove() and discard()

Display all courses and total number of courses using len()

```python
# creating a set

set = {"Python", "SQL", "Power BI", "Python", "Excel"}
print(" Original courses:", set)


#add a course

set.add("Tableau")
print("\n New_courses:" , set)

#update

set.update(["Powerpoint", "Looker"])
print("\n Updated_courser:" , set)

#remove
set.remove("Powerpoint")
print("\n After remove:" , set)

#discard

set.discard("Excel")
print("\n After discard:" , set)


print("\n Length of the set:" , len(set))
```
