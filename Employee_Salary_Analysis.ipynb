# **Case Study – Employee Salary Analysis**

**Instructions**

Create a list called salary Store employee salaries inside the list

**Use Built-in Functions to:**

Find maximum salary

Find minimum salary

Find total salary

Find number of employees using len()

Use lambda with map() to:
Increase each salary by 10%

Use lambda with filter() to:
Display salaries greater than 50000

**Display**:

Original salary list

Updated salary list

High salary list

Maximum salary

Minimum salary

Total salary

```python
#creating a list
store_emp_salaries = [8000,9400,10500,13000,17000,21000,25000,40000,42000,55000]

#Find maximum salary
max_salary = max(store_emp_salaries)
print("Max_salary:",max_salary)

#Find minimum salary
min_salary = min(store_emp_salaries)
print("Min_salary:",min_salary)

#Find total salary
total_salary = sum(store_emp_salaries)
print("Total_salary:",total_salary)

#Find number of employees using len()
no_of_emp = len(store_emp_salaries)
print("No_of_emp:",no_of_emp)

#Use lambda with map() to: Increase each salary by 10%

inc_salary = list(map(lambda s: s + (s*0.1),store_emp_salaries))
print("Updated salary:", inc_salary)

#Use lambda with filter() to: Display salaries greater than 50000

high_salary = list(filter(lambda s : s > 50000, inc_salary))
print("high_salary: ", high_salary)
```
