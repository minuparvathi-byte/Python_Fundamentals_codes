# Case Study – Password Verification System
Instructions:

Create a variable named password with a default password value.

Take password input from the user using input().

Use a while loop to: • Check whether the entered password matches the original password • Display "Incorrect password... try again" if password is wrong • Ask user to re-enter password until correct password is entered

Display "Access granted" when the correct password is entered

```python
# Case Study – Password Verification System

# Original password
password = "python123"

# Taking input from user
user_password = input("Enter the password: ")

# while loop for password checking
while user_password != password:
    print("Incorrect password... try again")
    user_password = input("Enter the password: ")

else:
    print("Access granted")
```

# Case Study – ATM PIN Verification System
Instructions:

Create a variable named correct_pin with a default PIN number.

Use a while loop to repeatedly ask the user to enter the PIN.

Check whether the entered PIN matches the correct PIN.

If the PIN is correct: • Display "Login Successful"
• Stop the loop using break

Otherwise: • Display "Incorrect PIN... Try Again"

```python
correct_pin = "756"


while True:
  user_pin = input("Enter your pin: ")

  if user_pin == correct_pin:
    print("Login Successful")
    break
  else:
    print("Incorrect pin... Try again")
```
