Case Study – Email Username Generator
Instructions:

• Create variables: full_name, domain_name

• Take input from the user using input() function

• Convert full_name into lowercase using lower()

• Remove spaces in name and replace with dot using replace()

• Split full_name using split() function

• Join name parts using join() function if needed

• Create email in format: name@domain

• Display only Full Name and Generated Email

• Use print() and f-string for output display

"""

```python
#taking input from user
Full_name =input("Enter Your Full_name:")
Domain_name= input("Enter Your Domain_name:")


# string operations
lower_name =Full_name.lower()
Email_name = lower_name.replace(" ",".")


#Create email in format
Email_id = Email_name + "@" + Domain_name + ".com"


#Display only Full Name and Generated Email
print("\n------user details------")

print(f"Full Name: {Full_name.upper()}")
print(f"Generated Email Id: {Email_id}")
```

Case Study – Student ID Card Generator
Instructions:

• Create variables: student_name, roll_number, department

• Take input from the user using input() function

• Convert student_name and department into uppercase using upper()

• Extract first 4 letters of student_name and department using slicing

• Create Student ID by combining name, department, and roll number

• Display Name, Department, Roll Number, and Student ID only

• Use print() and f-string for output display
"""

```python
# Student ID Card Generator
#taking input from students

student_name = input("Enter your name:")
roll_num = input("Enter your Roll_number:")
Department = input("Enter your Department:")

#string operations

Student_name= student_name.upper()
Departments= Department.upper()

first_4_name = Student_name[:5]
first_4_dept = Departments[:4]

#students_id creation

Student_id = first_4_name +"_" + first_4_dept + "_" + roll_num

#displaying the details

print("\n------Student Details------")

print(f"Name: {Student_name}")
print(f"Department: {Departments}")
print(f"Roll Number: {roll_num}")
print(f"Student ID: {Student_id}")

```
